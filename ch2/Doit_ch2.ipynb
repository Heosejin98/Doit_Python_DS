```python
# 실습 2-2
# 시퀀스 원소의  최대값  출력  하기
from typing import Any, Sequence

def max_of(a : Sequence) -> Any:
    maxinum = a[0]
    for i in range (1, len(a)):
        if(a[i] > maxinum):
            maxinum = a[i]
    return maxinum

if __name__ == '__main__':
    print('배열 최대 값')
    num = int(input('원소 갯수 입력' ))
    x = [None] * num
   
    for i in range(num):
        x[i] = int(input(f'x[{i}]값을 입력 :'))

    print(f'최대값 {max(x)}')
```

```python
# 실습 2-7
# 10진수 정숫값을 입력받아 2~36진수로 출력

def cord_conv(x: int, y: int) -> str:
    d = ''
    dchar = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'

    while x > 0:
        d += dchar[x % y]
        x = x // y

    return d[::-1] #역순으로 리턴

if __name__ == '__main__':
    print('10진수를 n진수로 변환')

    while True:
        while True:
            no = int(input('변환할 값 입력 :'))
            if no > 0:
                break
        while True:
            base = int(input('변환 진수 입력 : '))
            if 2 <= base <= 36:
                break
    
        print(f'{base}진수는 {cord_conv(no, base)}입니다')

        retry = input("한 번 더 변환 (Y : 예/ N : 아니요): ")
        if retry in {'N', 'n'}:
            break
```

```python
# 2C - 5
# 함수 사이에 인수 주고 받기

def sum(n):

    s = 0
    while n > 0:
       s += n 
       n -= 1
    return s

x = int(input('n의 값입력 :'))
print(f'1 ~ {x}사이의 합 {sum(x)}')
print(f'x의 값 {x}')
```

```python
# 실습 2-9
# 1000이하의 소수 나열 (알고리즘 개선)

cnt = 0
ptr = 0
prime = [None] * 500

prime[ptr] = 2
ptr += 1

for n in range(3, 1001, 2):
    for i in range(1, ptr):
        cnt += 1
        if n % prime[i] == 0 : 
            break
    else :
        prime[ptr] = n
        ptr += 1

for i in range(ptr):
    print(prime[i])
    
print(f'총 연산 횟수 : {cnt}')
```

```python
# 실습 2-9
# 1000이하의 소수 나열 (알고리즘 개선2) 제곱근
#
# 
cnt = 0
ptr = 0
prime = [None] * 500

prime[ptr] = 2
ptr += 1

prime[ptr] = 3
ptr += 1

for n in range(5, 1001, 2):
    i = 1
    while prime[i] * prime[i] <= n:
        cnt += 2
        if n % prime[i] == 0:
            break
        i += 1
    else :
        prime[ptr] = n
        ptr += 1
        cnt += 1

for i in range(ptr):
    print(f'print {i + 1} : {prime[i]}')
print(f'총 연산 횟수 : {cnt}')
```
