```python
# 3-3
# 선형 알고리즘 보초법으로 수정

from typing import Any, Sequence
from enum import Enum
import copy



def seq_search(li : Sequence, key : Any ) -> int :
    temp_li = copy.deepcopy(li)
    temp_li.append(key)

    i = 0
    while True:
        if temp_li[i] == key:
            break
        i += 1
    return -1 if i == len(li) else i

if __name__ == '__main__' :
    num = int(input('원소수를 입력 :'))
    li = [None] * num

    for i in range(num) :
        li[i] = i
    print(li)    
    
    ky = int(input('검색할 값을 입력 : '))

    idx = seq_search(li, ky)

    if idx == -1 :
        print('검색 결과가 없습니다')
    else :
        print(f'검색 결과 {li[idx]}은 {idx + 1}번째에 있습니다')
```

```python
# 3-4
# 이진 검색 알고리즘

from typing import Sequence, Any

def bin_search(a : Sequence, key : Any) -> int:
    center_idx = (0 + len(a)) // 2

    while True :
        if a[center_idx] == key :
            return [center_idx]
        elif a[center_idx] < key :
            center_idx += 1
        else:
            center_idx -= 1
        if center_idx == 0 or center_idx == len(a):
            break
    return -1

if __name__ == '__main__' :
    num = int(input('원소수를 입력 :'))
    li = [None] * num

    for i in range(num) :
        li[i] = i
    print(li)    
    
    ky = int(input('검색할 값을 입력 : '))

    idx = seq_search(li, ky)

    if idx == -1 :
        print('검색 결과가 없습니다')
    else :
        print(f'검색 결과 {li[idx]}은 {idx + 1}번째에 있습니다')
```

```python
# 3-5
# 체인법으로 해시 함수 구하기

from __future__ import annotations
from typing import Any, Type
from enum import Enum
import hashlib

class Node:
    """해시 구성 노드"""
    def __init__(self, key: Any, value: Any, next: Node) -> None:
        self.key = key #index
        self.value = value #원소값
        self.next = next #다음 주소

class ChainedHash:
    def __init__(self, capactiy: int) -> None:
        self.capacity = capactiy #해시 테이블의 크기 (key % [capactiy])
        self.table = [None] * self.capacity 

    def hash_value(self, key : Any) -> int: 
        if isinstance(key, int): #자료형 검사
            return key % self.capacity #해시값 구하기
        return (int(hashlib.sha256(str(key).encode()).hexdigest(), 16) % self.capacity)
    
    def search(self, key: Any) -> Any:
        hash = self.hash_value(key) 
        p = self.table[hash] #해시값에 맞는 노드를 주목

        while p is not None: 
            if p.key == key: 
                return p.value
            p = p.next 
        
        return None

    def add(self, key:Any, value: Any) -> bool:
        hash = self.hash_value(key)
        p = self.table[hash]

        #추가 실패시 다음 노드 주목
        while p is not None: 
            if p.key == key:
                return False
            p = p.next
        
        temp = Node(key, value, self.table[hash])
        self.table[hash] = temp #노드 추가
        return True
    
    def remove(self, key: Any) -> bool:
        hash = self.hash_value(key)
        p = self.table[hash] #해쉬키 값에 맞는 노드 주목
        pp = [None]

        while p is not None:
            if p.key == key: #해시테이블의 key와 입력한 key가 같다면
                if pp is None: #제일 처음 찾은경우
                    self.table[hash] = p.next 
                else : #삭제할 노드와 연결을 끊고 다음 노드를 주목하도록함
                    pp.next = p.next
                return True
            pp = p #현재 노드 가지고 있기
            p = p.next #다음 노드
        return False
    
    def dump(self) -> None: #출력
        print()
        for i in range(self.capacity):
            p = self.table[i]
            print(i, end='')
            while p is not None:
                print(f' -> {p.key} ({p.value})', end="")
                p = p.next
            print()

if __name__ == '__main__' :
    Menu = Enum('Menu', ['추가', '삭제', '검색', '덤프', '종료'])  # 메뉴를 선언

    def select_menu() -> Menu:
        """메뉴 선택"""
        s = [f'({m.value}){m.name}' for m in Menu]
        while True:
            print(*s, sep = '   ', end='')
            n = int(input(': '))
            if 1 <=  n <=  len(Menu):
                return Menu(n)

    hash = ChainedHash(13)     # 크기가 13인 해시 테이블을 생성

    while True:
      
        menu = select_menu()   # 메뉴 선택

        if menu == Menu.추가:  # 추가
            key = int(input('추가할 키를 입력하세요.: '))
            val = input('추가할 값을 입력하세요.: ')
            if not hash.add(key, val):
                print('추가를 실패했습니다!')

        elif menu == Menu.삭제:  # 삭제
            key = int(input('삭제할 키를 입력하세요.: '))
            if not hash.remove(key):
                print('삭제를 실패했습니다!')

        elif menu == Menu.검색:  # 검색
            key = int(input('검색할 키를 입력하세요.: '))
            t = hash.search(key)
            if t is not None:
                print(f'검색한 키를 갖는 값은 {t}입니다.')
            else:
                print('검색할 데이터가 없습니다.')

        elif menu == Menu.덤프:  # 덤프
            hash.dump()

        else:  # 종료
            break
        print()
```

```python
# 3-7
# 오픈 주소법으로 해시함수 구하기
# 주피터에서 잘 안돌아감

from __future__ import annotations
from typing import Any, Type
from enum import Enum
import hashlib

class Status(Enum):
    OCCUPIED = 0
    EMPTY = 1
    DELETE = 2

class Bucket:
    def __init__(self, key: Any = None, value: Any = None, 
                    stat : Status = Status.EMPTY) -> None:
        self.key = key
        self.value = value
        self.stat = stat
    
    def set(self, key: Any, value: Any, stat: Status) -> None:
        self.key = key
        self.value = value
        self.stat = stat

    def set_status(self, stat: Status) -> None:
        self.stat = stat

class Openhash:
    def __init__(self, capactiy: int) -> None:
        self.capactiy = capactiy
        self.table = [Bucket()] * self.capactiy

    def hash_value(self, key: Any) -> int:
        if isinstance(key, int):
            return key % self.capactiy
        return(int(hashlib.md5(str(key).encode).hexdigest(), 16) % self.capactiy)

    def rehash_value(self, key: Any) -> int:
        return(self.hash_value(key) + 1) % self.capactiy

    def search_node(self, key: Any) -> Any:
        hash = self.hash_value(key)
        p = self.table[hash]

        for i in range(self.capactiy):
            if p.stat ==  Status.EMPTY:
                break
            elif p.stat == Status.OCCUPIED and p.key == key:
                return p
            hash = self.rehash_value(hash)
            p = self.table[hash]
        return None

    def search(self, key: Any) -> Any:
        p = self.search_node(key)
        if p is not None:
            return p.value
        else:
            return None

    def add(self, key: Any, value: Any) -> bool:
        if self.search(key) is not None:
            return False

        hash = self.hash_value(key)
        p = self.table[hash]
        for i in range(self.capactiy):
            if p.stat == Status.EMPTY or p.stat == Status.DELETE:
                self.table[hash] = Bucket(key, value, Status.OCCUPIED)
                return True
            hash = self.rehash_value(hash)
            p = self.table[hash]
        return False

    def remove(self, key: Any) -> int:
        p = self.search_node(key)
        
        if p is None:
            return False
        p.set_status(Status.DELETE)
        return True

    def dump(self) -> None:
        for i in range(self.capactiy):
            print(f'{i:2} ', end='')
            if self.table[i].stat == Status.OCCUPIED:
                print(f'{self.table[i].key} ({self.table[i].value})')
            elif self.table[i].stat == Status.EMPTY:
                print('---미등록---')
            elif self.table[i].stat == Status.DELETE:
                print('---삭제완료---')


if __name__ == '__main__' :
    Menu = Enum('Menu', ['추가', '삭제', '검색', '덤프', '종료'])

    def select_menu() -> Menu:
        """메뉴 선택"""
        s = [f'({m.value}){m.name}' for m in Menu]
        while True:
            print(*s, sep = '  ', end='')
            n = int(input(': '))
            if 1 <=  n <= len(Menu):
                return Menu(n)

    hash = Openhash(13)  # 크기가 13인 해시 테이블 생성

    while True:
        menu = select_menu()  # 메뉴 선택

        if menu == Menu.추가:  # 추가
            key = int(input('추가할 키를 입력하세요.: '))
            val = input('추가할 값을 입력하세요.: ')
            if not hash.add(key, val):
                print('추가를 실패했습니다!')

        elif menu == Menu.삭제:  # 삭제
            key = int(input('삭제할 키를 입력하세요.: '))
            if not hash.remove(key):
                print('삭제를 실패했습니다!')

        elif menu == Menu.검색:  # 검색
            key = int(input('검색할 키를 입력하세요.: '))
            t = hash.search(key)
            if t is not None:
                print(f'검색한 키를 갖는 값은 {t}입니다.')
            else:
                print('검색할 데이터가 없습니다.')

        elif menu == Menu.덤프:  # 덤프
            hash.dump()

        else:  # 종료
            break
```
